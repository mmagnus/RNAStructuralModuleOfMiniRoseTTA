```python
import torch
from torch import tensor
import numpy as np

def rigidFrom3PointsTT(x1, x2, x3, mode = 0):
    """
    with no normalization
    """
    v1 = x3 - x2 # 0
    if mode: # 1
        v1 = x2 - x3
    v2 = x1 - x2

    e1 = v1 / torch.linalg.norm(v1)
    u2 = e1 * torch.dot(v2, e1)
    e2 = v2 - u2
    e2 = e2 / np.linalg.norm(e2)
    e3 = torch.cross(e1, e2) #.reshape((3,)), e2.reshape((3,))).reshape((3,1))
    R = torch.stack((e1,e2,e3), dim=-1)
    t = x2
    return R, t, e1, e2, e3

def get_ideal_coords(x1, x2, x3, x4 = torch.empty(0), mode = 0):
    R, t, e1, e2, e3 = rigidFrom3PointsTT(x1, x2, x3, mode) 
    if len(x4):
        X = torch.stack([x1, x2, x3, x4], dim=0) 
    else:
        X = torch.stack([x1, x2, x3], dim=0) 
    IC = np.round((X - t) @ R, decimals=3)
    return IC

OP2 = [48.880  ,77.258  ,41.111]
OP1 = [50.150,  76.113 , 39.198 ]
P = [50.001 , 77.254 , 40.137]

x1 = torch.tensor(OP1)
x2 = torch.tensor(P)
x3 = torch.tensor(OP2)
                  
get_ideal_coords(x1, x2, x3)              
```
